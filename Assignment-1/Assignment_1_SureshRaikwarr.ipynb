```python
r=102303163
ar=0.05*(r % 7)
br=0.3*((r%5)+1)
```

```python
import pandas as pd
import numpy as np
data=pd.read_csv("/kaggle/input/india-air-quality-data/data.csv", encoding="latin1")
```

```python
print(data.columns)
```

```python
x=data["no2"]
x
```

```python
print("Total values:", len(x))
print("NaN values:", x.isna().sum())
```

```python
x = x.dropna()
```

```python
x
```

```python
print("NaN values:", x.isna().sum())
```

```python
z=x+(ar * np.sin(br * x))
```

```python
z
```

```python
print("NaN values:", z.isna().sum())
```

```python
m=np.mean(z)
m
```

```python
v=np.var(z)
v
```

```python
l=1/(2*v)
l
```

```python
c = np.sqrt(l / np.pi)
c
```

```python
p = c * np.exp(-l * (z - m)**2)
p
```
